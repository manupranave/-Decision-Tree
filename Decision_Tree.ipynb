**4. Decision Tree Classification:**

```python
# Spliting the data into features(x) and target variable (y) .
x=df.drop(['sex','cp','trestbps','chol','fbs','restecg','thalch','exang','oldpeak','slope','thal','num'],axis=1)
y=df['age']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
from re import X
# Spliting the into training and testing sets .
# Training=80% .
# Testing=20% .
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
x,y
```

```python
# Fitting decision tree classifier to the training set .
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
classifier_dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
classifier_dt.fit(x_train,y_train)
```

```python
# Predicting the test set result .
y_pred1=classifier_dt.predict(x_test)
y_pred1
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
# Creating the confusion matrix .
cm=confusion_matrix(y_test,y_pred1)
clr=classification_report(y_test,y_pred1)
print(clr)
```

**5. Hyperparameter Tuning:**

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Defining hyperparameters grid .
param_grid={'max_depth':[3,5,7,10],'min_samples_split':[2,5,10],'criterion':['gini','entropy']}

# Creating a instance of gridsearchcv .
grid_serch=GridSearchCV(estimator=classifier_dt,param_grid=param_grid,cv=5)

# Fitting gridsearchcv to the training data .
grid_serch.fit(x_train,y_train)

# Accessing the best parameters and best estimator .
bst_param=grid_serch.best_params_
bst_estmr=grid_serch.best_estimator_

print('BEST PARAMETERS :',bst_param)
print('BEST ESTIMATOR :',bst_estmr)
```

**6. Model Evaluation and Analysis:**

```python
# Calculating metrics .
print('ACCURACY :',accuracy_score(y_test,y_pred1))
print('CONFUSION MATRIX :\n',confusion_matrix(y_test,y_pred1))
print('CLASSIFICATION REPORT :\n',classification_report(y_test,y_pred1))
```

2. What is the difference between the Label encoding and One-hot encoding?

• Label Encoding and One-Hot Encoding are both techniques used to connect categorical data into a numerical format that machine learning algorithms can understand. However, they work differently and are suitable for different types of categorical variables .

**• Label Encoding** assigns a unique integer to each category in a categorical variable. It works well for ordinal categorical variables, where there is a natural order or ranking among the categories .

**• One-Hot Encoding** creates binary columns for each category in a categorical variable. It works well for nominal categorical variables, where there is no inherent order among the categories .
